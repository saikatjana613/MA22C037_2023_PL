```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euclidean_distance(point1, point2):
    return math.sqrt((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2)
```

```python
def k_nearest_neighbors(points, k):
    n = len(points)
    pairwise_distances = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            pairwise_distances[i, j] = euclidean_distance(points[i], points[j])

    np.fill_diagonal(pairwise_distances, np.inf)
    
    k_nearest_indices = np.argpartition(pairwise_distances, k, axis=1)[:, :k]

    return k_nearest_indices
```

```python
P = np.array([[0,0], [0.2, 0.22], [0.1, -0.1], [1,1], [1.1, 0.9], [0.8, 0.9], [1.1, 0.63], [0.58, -0.1], [0.63, 0.1], [0.67, -0.3], [0.8,-0.23], [0.8, 0.6]])

k = 3

k_nearest_indices = k_nearest_neighbors(P, k)
```

```python
plt.scatter(P[:, 0], P[:, 1])

for i in range(len(P)):
    nearest_points = P[k_nearest_indices[i]]
    for neighbor_point in nearest_points:
        plt.arrow(P[i, 0], P[i, 1], neighbor_point[0] - P[i, 0], neighbor_point[1] - P[i, 1], head_width=0.02, head_length=0.03)

plt.xlabel('X')
plt.ylabel('Y')

plt.grid(True)
plt.show()
```
