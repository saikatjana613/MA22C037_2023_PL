```python
import pandas as pd
%matplotlib inline
```

```python
import yfinance as yf

# Define the list of ticker symbols
symbols = ['AAPL', 'GOOG', 'TSLA']

# Fetch data from Yahoo Finance
s = yf.download(symbols, start='2023-01-01')
```

```python
# â€¢ Find out if a heirarchical indexing (MultiIndex) is being used in this data.

is_multiindex = isinstance(s.index, pd.MultiIndex)
print("Is MultiIndex being used in this data?", is_multiindex)
```

```python
# â€¢ Access and plot the closing price of AAPL on all days in the data.

import matplotlib.pyplot as plt

aapl_closing = s['Adj Close']['AAPL']
aapl_closing.plot(title='AAPL Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

```python
# â€¢ Print the closing price of all three stocks yesterday. 

# Find available dates in the index
available_dates = s.index.get_level_values(0).unique()

# Print available dates
print("Available dates in the DataFrame:")
print(available_dates)

# Select one of the available dates to check
date_to_check = '2023-10-06'  # Replace with an available date from the list

# Access and print the closing prices for the selected date
closing_prices = s.loc[date_to_check, 'Adj Close']
print("Closing prices of all three stocks on", date_to_check)
print(closing_prices)
```

```python
# â€¢ Extract a smaller data frame with no MultiIndex containing only TSLA data. 

tsla_data = s['Adj Close']['TSLA'].reset_index()
print("\nDataFrame with TSLA data only:")
print(tsla_data)
```
