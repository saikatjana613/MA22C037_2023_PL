```python
import numpy as np
```

```python
def matrix(n, k):
    A = np.zeros((n, n))

    for i in range(n):
        for j in range(max(0, i - k), min(n, i + k + 1)):
            A[i, j] = 1
    return A


n = 6
k = 2
A = matrix(n, k)
print(A)
```

```python
def average_matrix(n, k):
    A = matrix(n,k)
    
    for i in range(n):
        row_sum = sum(A[i, :])
        A[i, :] /= row_sum

    return A

n = 20
k = 2
A = average_matrix(n, k)
print(A)
```
