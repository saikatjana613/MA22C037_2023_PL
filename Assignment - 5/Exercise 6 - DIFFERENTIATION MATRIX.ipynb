```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the size of the matrix and grid spacing
n = 1000  # Choose the number of grid points
h = 2 * np.pi / (n - 1)  # Grid spacing


a=np.diag(np.ones(n - 1),k=1)
b=np.diag(-np.ones(n ),k=0)

# Create the differentiation matrix D
D = np.delete(a+b,n-1, 0)

# Define the function f(x) = sin(x) for equally spaced x values
x = np.linspace(0, 2*np.pi, n)

# f_x = np.sin( x)
f_x=np.sin(x)

# Compute the derivative using the differentiation matrix D
df = np.dot(D, f_x/h)

#actual derivative of sin(x)
y=np.cos(x)
```

```python
y=np.cos(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x[:-1], df,'--', label="Approx. Derivative",color='blue')
plt.plot(x, y,':',label='cos(x)', color='red')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Approximate Derivative of f(x) =sin(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the function f(x) = x^2 + 2sin(10x) for equally spaced x values
f_x_new = x**2 + 2 * np.sin(10 * x)

# Compute the derivative using the differentiation matrix D
df_new = np.dot(D, f_x_new/h)

y1=2*x+20*np.cos(10*x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x[:-1], df_new, label="Approx. Derivative")
plt.plot(x, y1,':',label='2x+20cos(10x)', color='red')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Approximate Derivative of f(x) = x^2 + 2sin(10x)")
plt.grid(True)
plt.legend()
plt.show()
```
