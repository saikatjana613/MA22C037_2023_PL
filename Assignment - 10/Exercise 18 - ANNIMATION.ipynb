```python
import numpy as np
```

```python
# Define the function f(x, t)
def f(x, t):
    return 0.5 * (x + t)**2 + 2 * np.sin(10 * (x - t))
```

```python
# Task 1: Use matplotlib.animation module to display changes in the plot of f over x with respect to time t. 

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Create a figure and axis for plotting
fig, ax = plt.subplots()
x_values = np.linspace(0, 10, 1000)
line, = ax.plot(x_values, f(x_values, 0))  # Initialize the plot with t = 0

# Define the animation function
def animate(t):
    ax.clear()
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 150)  # Adjust the y-axis limits if needed
    line, = ax.plot(x_values, f(x_values, t))
    ax.set_title(f'f(x, t) for t = {t:.1f}')

# Create the animation
animation = FuncAnimation(fig, animate, frames=np.linspace(0, 10, 100), interval=50)

# Display the animation
plt.show()
```

```python
# Task 2: Use the celluloid module to perform the same task. 

import matplotlib.pyplot as plt
from celluloid import Camera

# Create a figure and axis for plotting
fig, ax = plt.subplots()
camera = Camera(fig)

# Iterate over a range of t values
for t in np.linspace(0, 10, 100):
    x_values = np.linspace(0,10, 1000)
    ax.plot(x_values, f(x_values, t))
    ax.set_title(f'f(x, t) for t = {t:.1f}')
    camera.snap()

# Create the animation
animation = camera.animate(interval=50)

# Display the animation
plt.show()
```

```python
#  Which is faster?
#  Ans: matplotlib.animation is typically faster as it directly interacts with the animation backend of Matplotlib.
    
#  Which is  more convenient? 
#  Ans: celluloid is more convenient and user-friendly for simple animations. 
 #   It simplifies the process and reduces the amount of code needed compared to using matplotlib.animation directly
```

```python
# Task 3: Add a text labeling the time at each snapshot that the animation is composed of

import matplotlib.pyplot as plt
from celluloid import Camera
import numpy as np

def f(x, t):
    return 0.5 * (x + t)**2 + 2 * np.sin(10 * (x - t))

# Create a figure and axis for plotting
fig, ax = plt.subplots()
camera = Camera(fig)

# Iterate over a range of t values
for t in range(10):
    x_values = np.linspace(0, 10, 1000)
    ax.plot(x_values, f(x_values, t))
    ax.annotate(f'Time = {t:.2f}', xy=(5, 3), fontsize=12)
    camera.snap()

# Create the animation
animation = camera.animate()

# Display the animation
plt.show()
```
