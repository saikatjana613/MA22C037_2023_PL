```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn; seaborn.set();
from scipy.integrate import solve_ivp
import numpy as np
```

```python
def seir_f(t, y, alpha, beta, gamma, delta):
    r, x = y
    return np.array([alpha*r-beta*r*x,delta*r*x-gamma*x])
```

```python
# try some parameter values
alpha = 1.1
beta = 0.4
gamma = 0.1
delta = 0.1
```

```python
sol = solve_ivp(seir_f, [0, 70], [5, 2],args=(alpha, beta, gamma, delta),rtol=1e-6)
```

```python
# Task-1
```

```python
fig = plt.figure(); ax = fig.gca()
curves = ax.plot(sol.t, sol.y.T)
plt.title('Phase Portrait of Rabbit and Fox Population Dynamics')
ax.legend(curves, ['rabbits(r)', 'foxes(x)']);
```

```python
# Task -2 
```

```python
# Generate 10 random initial conditions for x and r
num_conditions = 10
x_initial_conditions = np.random.uniform(1, 9, num_conditions)
r_initial_conditions = np.random.uniform(1, 9, num_conditions)

# Loop through the initial conditions and generate phase portraits
for x0, r0 in zip(x_initial_conditions, r_initial_conditions):
    sol = solve_ivp(seir_f, [0, 70], [r0, x0], args=(alpha, beta, gamma, delta), rtol=1e-6)

    fig = plt.figure()
    ax = fig.gca()
    curves = ax.plot(sol.t, sol.y.T)
    plt.title('Phase Portrait of Rabbit and Fox Population Dynamics')
    ax.legend(curves, ['rabbits(r)', 'foxes(x)'])

plt.show()
```

```python
equilibria = [
    [0, alpha / beta],  # x values
    [0, gamma / delta]  # r values
]

# Plot the phase portrait
plt.figure(figsize=(8, 8))

num_conditions = 10
x_initial_conditions = np.random.uniform(1, 9, num_conditions)
r_initial_conditions = np.random.uniform(1, 9, num_conditions)

# Loop through the initial conditions and generate phase portraits
for x0, r0 in zip(x_initial_conditions, r_initial_conditions):
    sol = solve_ivp(seir_f, [0, 70], [r0, x0], args=(alpha, beta, gamma, delta), rtol=1e-6)

    plt.plot(sol.y[1], sol.y[0], label=f'x0={x0}, r0={r0}')

# Plot the equilibria
plt.scatter(equilibria[0], equilibria[1], color='black', marker='o', label='Equilibria')
plt.xlabel('Foxes (x)')
plt.ylabel('Rabbits (r)')
plt.legend()
plt.title('Phase Portrait with Equilibria')
plt.grid(True)
plt.show()
```
